#**Explicaci√≥n sobre Pruebas de Escritorio en Programaci√≥n**

**¬øQu√© es una prueba de escritorio?**

Una prueba de escritorio es una t√©cnica utilizada en programaci√≥n para simular la ejecuci√≥n de un algoritmo o c√≥digo sin necesidad de una computadora. Consiste en analizar l√≠nea por l√≠nea c√≥mo se comporta el programa utilizando l√°piz y papel, con el objetivo de detectar errores l√≥gicos antes de ejecutarlo en un entorno real.

**¬øC√≥mo realizar una prueba de escritorio?**

Para hacer una prueba de escritorio correctamente, se sigue el siguiente procedimiento:

1.	Escribir el c√≥digo o algoritmo que se quiere probar.
2.	Identificar las variables y sus valores iniciales.
3.	Dibujar una tabla con las variables que se van a modificar durante la ejecuci√≥n.
4.	Ejecutar el c√≥digo l√≠nea por l√≠nea, registrando los cambios en la tabla.
5.	Observar si el resultado obtenido es el esperado o si hay errores.
